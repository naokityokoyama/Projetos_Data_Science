## **CAMADAS**

*    A camada **flatten** (achatada) serve para alterarmos a organizao de um   dado, mudando um formato de matriz para um formato de vetor. Ela no mexe nas conexes entre as unidades da rede.

*    A camada **dense** (densa)  aquela que faz as conexes entre todas as unidades da camada 2 com as unidades da camada 1.

*    Usando a camada **croppin**g (cortadora) iremos cortar fora ou ignorar algumas partes dos dados. Ao fazer isso, ela no vai alterar a quantidade de conexes que temos na rede.

*    A camada **reshape** (remodeladora)  usada para mudarmos a forma de um dado, ento podemos fazer, por exemplo, com que uma tupla se transforme em um array. Mas ao usarmos essa camada, no alteramos as conexes das nossas unidades.


```
# no keras, deep learning  diferente do ML precisamos definir o modelo, dados de entrada, processamento e saida
#modelo keras.Sequential  #sempre sequencial
#  entrada-> keras.layers#camadas .Flatten #achatamento (input_shape(28, 28)) 28x28 
#  processamento -> keras.layers. Dense (256, activation=funcao) #densidade, a quantidade(256) de camadas qua vao se conectar com a entrada / ativando a funcao para se conectar (existe varias dentro do tensoflow)
                                                                 # EX activarion = tensorflow.nn.relu (dentro do tenso funcao nn network neural , relu(que tem outras tb ))
#  saida -> keras.layers.Dense(10,activation=tensorflow.nn.softmax) 10-> so as quantidades de tipos de classificao
 
 
 #funcao : relu (funcao no tem negativo e no linear relu-> unidade linear identificada)    
          #softmax (probabilidade)                                               
```

```
#modelo (modelo , compilao e treino)

#modelo
modelo = keras.Sequential([
  keras.layers.Flatten(input_shape=(28,28)),     #variavel input_shape = 28,28                
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dense(10, activation=tensorflow.nn.softmax)
])

#compitaao (optimizador adam  um famoso optimzador e a perda)
modelo.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

#treinando modelo
modelo.fit(imagens_treino, identificacao_treino)
```

```
#validao (treino, validacao e teste)
imagens_treino = imagens_treino/float(255) 

modelo = keras.Sequential([
  keras.layers.Flatten(input_shape=(28,28)),                    
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dense(10, activation=tensorflow.nn.softmax)
])

modelo.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

historico = modelo.fit(imagens_treino, identificacao_treino, epochs=5, validation_split=0.2) #validacao 
```

```
#tecnica dos neuronios "adormecidos" onde no vao receber a quanidade igual 
imagens_treino = imagens_treino/float(255) 

modelo = keras.Sequential([
  keras.layers.Flatten(input_shape=(28,28)),                    
  keras.layers.Dense(256, activation=tensorflow.nn.relu),
  keras.layers.Dropout(0.2), #20% adormecido 
  keras.layers.Dense(10, activation=tensorflow.nn.softmax)
])

modelo.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')

historico = modelo.fit(imagens_treino, identificacao_treino, epochs=15, validation_split=0.2)
```

```
plt.plot(historico.history['accuracy'])
plt.plot(historico.history['val_accuracy'])
plt.legend(['treino', 'valiacao'])
plt.xlabel('Epoca')
plt.ylabel('Acuracia')
plt.title('Acucaria por validacao ')
```

```
plt.plot(historico.history['loss'])
plt.plot(historico.history['val_loss'])
plt.legend(['treino', 'avaliacao'])
plt.xlabel('Epoca')
plt.ylabel('Perda')
plt.title('perda por validacao ')
```
